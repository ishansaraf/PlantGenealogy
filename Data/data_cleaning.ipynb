```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./PlantGenealogy/Data/strains.csv',sep=';',names=['name','description','grow info','effects','medical','negatives','flavor','parents','avg rating','num ratings'])
```

```python
df[pd.notnull(df['parents'])]
```

```python
flavors = ['Ammonia', 'Apple', 'Apricot', 'Berry', 'Blue Cheese', 'Blueberry', 'Butter', 'Cheese', 'Chemical', 'Chestnut', 'Citrus', 'Coffee', 'Diesel', 'Earthy', 'Flowery', 'Grape', 'Grapefruit', 'Honey', 'Lavender', 'Lemon', 'Lime', 'Mango', 'Menthol', 'Mint', 'Nutty', 'Orange', 'Peach', 'Pear', 'Pepper', 'Pine', 'Pineapple', 'Plum', 'Pungent', 'Rose', 'Sage', 'Skunk', 'Spicy/Herbal', 'Strawberry', 'Sweet', 'Tar', 'Tea', 'Tobacco', 'Tree', 'Fruit', 'Tropical', 'Vanilla', 'Violet', 'Woody']
effects = ['Anxious', 'Aroused', 'Creative', 'Energetic', 'Euphoric', 'Focused', 'Giggly', 'Happy', 'Hungry', 'Relaxed', 'Sleepy', 'Talkative', 'Tingly', 'Uplifted',]

```

```python
import re
def ef(sr, subject):
    if subject in sr:
        ar = re.split(",|:",sr)
        if subject in ar:
            pos = ar.index(subject)
            return float(ar[pos+1])
    return 0.0
```

```python
def fin(sr, medical):
    ar = re.split(",|:",sr)
    i = 0
    while i < len(ar):
        word = ar[i]
        if word not in medical:
            medical.append(word)
        i+=2
```

```python
medical = []
df.medical.apply(lambda x: fin(str(x),medical))
medical.remove('')
medical.remove('nan')
medical = medical[:-1]
medical
```

```python
def flav(sr, subject):
    if subject in sr:
        ar = re.split(",", sr)
        if subject in ar:
            pos = ar.index(subject)
            return 1.0
    return 0.0
```

```python
negatives = []
df.negatives.apply(lambda x: fin(str(x),negatives))
negatives.remove('')
negatives.remove('nan')
```

```python
effect_list = pd.DataFrame(index=df.index)
effect_list['name'] = df.name
for effect in effects:
    effect_list[effect] = df.effects.apply(lambda x: ef(str(x), effect))
effect_list = effect_list.select_dtypes(['number'])
flavor_list = pd.DataFrame()
flavor_list['name'] = df.name
for flavor in flavors:
    flavor_list[flavor] = df.flavor.apply(lambda x: flav(str(x), flavor))
flavor_list = flavor_list.select_dtypes(['number'])
medical_list = pd.DataFrame()
medical_list['name'] = df.name
for med in medical:
    medical_list[med] = df.medical.apply(lambda x: ef(str(x), med))
medical_list = medical_list.select_dtypes(['number'])

neg_list = pd.DataFrame()
neg_list['name'] = df.name
for neg in negatives:
    neg_list[neg] = df.negatives.apply(lambda x: ef(str(x), neg))
neg_list = neg_list.select_dtypes(['number'])

neg_list
```

```python
# concat new dataframes together
new_df = pd.DataFrame()
new_df['name'] = df.name
new_df['description'] = df.description
new_df['parents'] = df.parents
for i in range(1,effect_list.shape[1]):
    new_df[effect_list.columns[i]] = effect_list.iloc[:,i]
for i in range(1,flavor_list.shape[1]):
    new_df[flavor_list.columns[i]] = flavor_list.iloc[:,i]
for i in range(1,medical_list.shape[1]):
    new_df[medical_list.columns[i]] = medical_list.iloc[:,i]
for i in range(1,neg_list.shape[1]):
    new_df[neg_list.columns[i]] = neg_list.iloc[:,i]
df['avg rating'] = pd.to_numeric(df['avg rating'],errors='coerce')
df['num ratings'] = pd.to_numeric(df['num ratings'],errors='coerce')
df1 = df.select_dtypes(['number'])
new_df['avg rating'] = df1['avg rating']
new_df['num ratings'] = df1['num ratings']
new_df.fillna(value=0)
new_df['name'] = df.name
new_df['description'] = df.description
new_df['parents'] = df.parents
new_df
```

```python
new_df.to_csv('./strains_cleaned.csv',index=False)
```
